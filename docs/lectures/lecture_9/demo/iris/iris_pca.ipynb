### PCA with the iris data set

Preparing the Iris Dataset

For the following tutorial, we will be working with the famous "Iris" dataset that has been deposited on the UCI machine learning repository
(https://archive.ics.uci.edu/ml/datasets/Iris).

The iris dataset contains measurements for 150 iris flowers from three different species.

The three classes in the Iris dataset are:

- Iris-setosa (n=50)
- Iris-versicolor (n=50)
- Iris-virginica (n=50)

And the four features of in Iris dataset are:

- sepal length in cm
- sepal width in cm
- petal length in cm
- petal width in cm

## Loading the Dataset

In order to load the Iris data directly from the UCI repository, we are going to use the superb pandas library. 



```python
import numpy as np # linear algebra
import pandas as pd
from sklearn.decomposition import PCA
from sklearn import datasets

import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.style.use('seaborn')
%matplotlib inline
```

## Principal compomenet analysis

Import iris dataset and view as a dataframeÂ¶

```python
# import some data to play with
iris = datasets.load_iris()
df = pd.DataFrame(data=iris['data'],columns=iris['feature_names'])
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

## Describe Features

```python
sns.pairplot(data=df,kind='scatter', hue='species')
```

## Conclusion

Three features describes relally good the classes so we could reduce the number of features to 3

```python
reduced = PCA(n_components=3).fit_transform(df.iloc[:,0:4])
```

```python
# plot
fig = plt.figure(figsize=(18, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(reduced[:,0], reduced[:,1],reduced[:,2], c=iris.target, s=100)
ax.view_init(20, 90)
plt.show()
```
